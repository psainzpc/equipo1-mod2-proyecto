## CANCIONES POP 2019

```python
#LISTA DE CANCIONES POP 2019
dict_tracks_pop_2019 ={"canciones":[], "artistas": [], "género": "Pop", "tipo": "track", "año lanzamiento": []}
for offset in range(0,500,50):
    canciones_2019 = sp.search(q="genre:pop year:2019", type="track", limit=50, offset=offset)
    for i in range(0,50,1):
        dict_tracks_pop_2019["canciones"].append(canciones_2019["tracks"]["items"][i]["name"])
        

print(dict_tracks_pop_2019)    
len(dict_tracks_pop_2019)
```

### ARTISTAS POP 2019

```python
#ARTISTAS DE CANCIONES POP 2019

for offset in range(0, 500, 50):
    canciones = sp.search(q='genre:"pop" year:2019', type='track', limit=50, offset=offset)
    for track in canciones['tracks']['items']:
        dict_tracks_pop_2019["artistas"].append(track["artists"][0]['name'])
print(dict_tracks_pop_2019)
len(dict_tracks_pop_2019)
```

### AÑO LANZAMIENTO 2019

```python
#AÑO DE LANZAMIENTO
for offset in range(0, 500, 50):
    canciones = sp.search(q='genre:"pop" year:2019', type='track', limit=50, offset=offset)
    for track in canciones['tracks']['items']:
        dict_tracks_pop_2019["año lanzamiento"].append(track["album"]['release_date'])
print(dict_tracks_pop_2019)
len(dict_tracks_pop_2019)
```
