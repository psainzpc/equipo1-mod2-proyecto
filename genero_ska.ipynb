CANCIONES SKA 2019

```python
dict_tracks_ska_2019={"canciones":[],"artistas": [], "género": "ska", "tipo": "track", "año lanzamiento": []}
for offset in range (0,500,50):
    canciones_ska_2019 = sp.search(q='genre:"ska" year:2019', type="track", limit=50, offset=offset)
    for i in range(0,50,1):
        dict_tracks_ska_2019["canciones"].append(canciones_ska_2019["tracks"]["items"][i]["name"])
        
print(dict_tracks_ska_2019)
len(dict_tracks_ska_2019)
```

ARTISTAS SKA 2019

```python
for offset in range(0, 500, 50):
    canciones_ska_2019 = sp.search(q='genre:"ska" year:2019', type="track", limit=50, offset=offset)
    for track in canciones_ska_2019['tracks']['items']:
        dict_tracks_ska_2019["artistas"].append(track["artists"][0]['name'])
print(dict_tracks_ska_2019)
len(dict_tracks_ska_2019)
```

AÑO DE LANZAMIENTO

```python
for offset in range(0, 500, 50):
    canciones_ska_2019 = sp.search(q='genre:"ska" year:2019', type='track', limit=50, offset=offset)
    for track in canciones_ska_2019['tracks']['items']:
        dict_tracks_ska_2019["año lanzamiento"].append(track["album"]['release_date'])
print(dict_tracks_ska_2019)
len(dict_tracks_ska_2019)
```

```python
#ejemplo de como muestra los resultados:
result = musicbrainzngs.search_artists(artist='Inspector')
print(result)
```

```python
artistas_ska_2019_mb = {"artista": [], "pais de origen": [], "area de origen": [], "fecha inicio actividad": [], "fin de actividad": []} 
for a in lista_artistas_ska_2019:
    result = musicbrainzngs.search_artists(artist=a)
    try:
        artistas_ska_2019_mb["artista"].append(result["artist-list"][0]["name"])
    except:
        artistas_ska_2019_mb["artista"].append("no data")
    try:
        artistas_ska_2019_mb["pais de origen"].append(result["artist-list"][0]["area"]["name"])
    except:
        artistas_ska_2019_mb["pais de origen"].append("no data")
    try:
        artistas_ska_2019_mb["area de origen"].append(result["artist-list"][0]["begin-area"]["name"])
    except:
        artistas_ska_2019_mb["area de origen"].append("no data") 
    try:
        artistas_ska_2019_mb["fecha inicio actividad"].append(result["artist-list"][0]["life-span"]["begin"])
    except:
        artistas_ska_2019_mb["fecha inicio actividad"].append("no data")
    try:
        if result["artist-list"][0]["life-span"]["ended"] == False:
            artistas_ska_2019_mb["fin de actividad"].append("no activo")
        else:
            artistas_ska_2019_mb["fin de actividad"].append("activo")
    except:
        artistas_ska_2019_mb["fin de actividad"].append("no data")
```
